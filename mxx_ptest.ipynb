```python
'''
#103. 数组的插入和删除
问题描述
对数组执行一系列指令，输出操作后的结果

第一行输入列表中的数字或字符，空格间隔

第二行输入操作指令条数

第三行开始每一行是一个操作指令 +32表示在下标为3的位置插入字符'2' -23表示从下标为2的位置开始删除连续三个位置的值

第一行输出操作后的列表，不存在操作指令不可操作的情况。数组空间最大为10个元素。
'''
A=list(map(int,input().split()))
a=['!']*10
for i in range(len(A)):
    a[i]=A[i]
N=len(A)
b=int(input())
for i in range(b):
    m=input()
    if m[0]=='+':
        k=int(m[1])
        j=9
        while j>=k:
            a[j]=a[j-1]
            j=j-1
        a[k]=int(m[2])
        N=N+1
    if m[0]=='-':
        k=int(m[1])
        jj=k
        geshu=int(m[2])
        for x in range(N-geshu-k):
            a[jj]=a[jj+geshu]
            jj=jj+1
        N=N-geshu
final=''
for i in range(N):
    final+=str(a[i])+' '
print(final[:-1])
```

```python

'''
#107. 链表的删除
问题描述
对链表执行删除指令，输出操作后的结果

第一行输入链表中每一个节点的字符和指针，空格间隔，最后输入链表的head

第二行输入需要删除的数据key

第一行输出遍历链表节点的数据。

example 1:
Input
33 8 6 4 83 -1 48 9 26 0 56 2 43 7 44 3 38 6 53 5 1
48

Output
6 26 33 38 43 44 53 56 83

example 2:
Input
99 -1 5 2 26 6 47 9 75 5 93 0 40 3 54 8 56 4 52 7 1
8

Output
未找到

'''

a=list(map(int,input().split()))
key=int(input())
head=a[-1]
data=[]
wei=[]
for i in range(len(a)-1):
    if i%2==0:
        wei.append(a[i])
    if i%2==1:
        wei.append(a[i])
        data.append(wei)
        wei=[]
k=head
m=''
while k!=-1:
    m+=str(data[k][0])+' '
    k=data[k][1]
print(m[:-1])

weizhi=head  ##当前位置
last=head   ##记录前一个元素
while weizhi!=-1:
    if data[weizhi][0]==key:
        if last==head:         ##第一个元素被删除，需要特殊处理
            head=data[weizhi][1]
        else:
            data[last][1]=data[weizhi][1] #修改上一个元素的指针，跳过被删除元素
        break
    else:                      ##遍历
        last=weizhi
        weizhi=data[weizhi][1]
if weizhi!=-1:
    k=head
    m=''
    while k!=-1:
        m+=str(data[k][0])+' '
        k=data[k][1]
    print(m[:-1])
else:
    print("未找到")
```
